### 1- Das 3 Categorias que mais vendem, quais são os 10 clientes que mais gastaram em cada categoria de produto, e qual foi o valor total gasto por categoria?

```python
sql='''

with Top_customers_cama_mesa_banho as (
    SELECT
    
    p.product_category_name,
    cus.customer_id,
    SUM(op.payment_value) AS ValorGastoTotal
    FROM
    orders AS o
    JOIN customers AS cus ON o.customer_id = cus.customer_id
    JOIN order_payments AS op ON o.order_id = op.order_id
    JOIN order_items AS oi ON o.order_id = oi.order_id
    JOIN products AS p ON oi.product_id = p.product_id
    Where p.product_category_name == 'cama_mesa_banho'
    GROUP BY
    cus.customer_id,
    p.product_category_name
    ORDER BY
    ValorGastoTotal DESC
    limit 10
),
Top_customers_beleza_saude as (
    SELECT
    
    p.product_category_name,
    cus.customer_id,
    SUM(op.payment_value) AS ValorGastoTotal
    FROM
    orders AS o
    JOIN customers AS cus ON o.customer_id = cus.customer_id
    JOIN order_payments AS op ON o.order_id = op.order_id
    JOIN order_items AS oi ON o.order_id = oi.order_id
    JOIN products AS p ON oi.product_id = p.product_id
    Where p.product_category_name == 'beleza_saude'
    GROUP BY
    cus.customer_id,
    p.product_category_name
    ORDER BY
    ValorGastoTotal DESC
    limit 10 ),
    
Top_customers_informatica_acessorios as (
    SELECT
    
    p.product_category_name,
    cus.customer_id,
    SUM(op.payment_value) AS ValorGastoTotal
    FROM
    orders AS o
    JOIN customers AS cus ON o.customer_id = cus.customer_id
    JOIN order_payments AS op ON o.order_id = op.order_id
    JOIN order_items AS oi ON o.order_id = oi.order_id
    JOIN products AS p ON oi.product_id = p.product_id
    Where p.product_category_name == 'informatica_acessorios'
    GROUP BY
    cus.customer_id,
    p.product_category_name
    ORDER BY
    ValorGastoTotal DESC
    limit 10)
    
Select *

    from Top_customers_cama_mesa_banho as a
union all
    select * 
    from Top_customers_beleza_saude as b 
union all
    select *
    from Top_customers_informatica_acessorios as c






''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql.head(100)
```

### 2- Quais são os produtos que tiveram um aumento de preço superior a 20% nos últimos três meses de dados disponíveis?

```python
sql='''

with LastThreeMonths as (
SELECT oi.price,
        oi.product_id,
         max(o.order_purchase_timestamp) as max_date_purchase
        
    
    
    
    FROM
    orders AS o
    JOIN customers AS cus ON o.customer_id = cus.customer_id
    JOIN order_payments AS op ON o.order_id = op.order_id
    JOIN order_items AS oi ON o.order_id = oi.order_id
    JOIN products AS p ON oi.product_id = p.product_id
    Where o.order_purchase_timestamp >= date('2018-09-03 09:06:57', '-3 months')
    GROUP BY
    oi.price,
    oi.product_id
    ORDER BY o.order_purchase_timestamp
),   
BeforeLastThreeMonths as (
SELECT oi.price,
        oi.product_id,
         max(o.order_purchase_timestamp) as max_date_purchase
        
    
    
    
    FROM
    orders AS o
    JOIN customers AS cus ON o.customer_id = cus.customer_id
    JOIN order_payments AS op ON o.order_id = op.order_id
    JOIN order_items AS oi ON o.order_id = oi.order_id
    JOIN products AS p ON oi.product_id = p.product_id
    Where o.order_purchase_timestamp < date('2018-09-03 09:06:57', '-3 months')
    GROUP BY
    oi.price,
    oi.product_id
    ORDER BY o.order_purchase_timestamp
)

select count(distinct ltm.product_id)

from LastThreeMonths as ltm
inner join BeforeLastThreeMonths as bltm
on ltm.product_id = bltm.product_id
where ((ltm.price / bltm.price)-1) >= 0.20









''';

    
df_sql = pd.read_sql_query(sql,con=engine)
df_sql
```
